```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("../data/results.csv")
print("Data loaded successfully!")
```

```python
# 1. Number of matches
print("Total matches:", df.shape[0])

# 2. Date range
df['date'] = pd.to_datetime(df['date'])
print("Earliest date:", df['date'].min())
print("Latest date:", df['date'].max())
```

```python
# 3. Number of unique countries
total_teams = pd.concat([df['home_team'], df['away_team']]).unique()
print("Number of unique countries:", len(total_teams))

# 4. Team that appears most as home team
most_home = df['home_team'].value_counts().head(1)
print("\nTeam with most home matches:")
print(most_home)
```

```python
# Create total_goals column
df["total_goals"] = df["home_score"] + df["away_score"]

# 5. Average goals per match
print("5. Average goals per match:", df["total_goals"].mean())

# 6. Highest scoring match
max_goals_row = df.loc[df["total_goals"].idxmax()]
print("\n6. Highest scoring match:")
print(f"{max_goals_row['home_team']} {max_goals_row['home_score']} - {max_goals_row['away_score']} {max_goals_row['away_team']}")

# 7. Home vs Away goals
print("\n7. Total home goals:", df["home_score"].sum())
print("   Total away goals:", df["away_score"].sum())

# 8. Most common total goals
print("\n8. Most common total goals:", df["total_goals"].mode()[0])
```

```python
# Function to determine match result
def match_result(row):
    if row["home_score"] > row["away_score"]:
        return "Home Win"
    elif row["home_score"] < row["away_score"]:
        return "Away Win"
    else:
        return "Draw"

df["result"] = df.apply(match_result, axis=1)

# 9. Percentage of home wins
result_counts = df["result"].value_counts(normalize=True) * 100
print("9. Home wins percentage:", result_counts["Home Win"])

# 10. Does home advantage exist?
print("\n10. Match outcomes:")
print(result_counts)

# 11. Country with most wins historically
home_wins = df[df['result'] == 'Home Win']['home_team'].value_counts()
away_wins = df[df['result'] == 'Away Win']['away_team'].value_counts()
total_wins = home_wins.add(away_wins, fill_value=0).sort_values(ascending=False)

print("\n11. Country with most wins:")
print(total_wins.head(1))
```

```python
# Histogram of goals
plt.figure(figsize=(10,6))
df["total_goals"].hist(bins=15, edgecolor='black')
plt.title("Distribution of Goals Per Match")
plt.xlabel("Total Goals")
plt.ylabel("Number of Matches")
plt.show()

# Bar chart of match outcomes
plt.figure(figsize=(8,5))
df['result'].value_counts().plot(kind='bar', color=['green', 'red', 'gray'])
plt.title('Match Results Distribution')
plt.xlabel('Result')
plt.ylabel('Count')
plt.show()

# Top 10 teams by total wins
plt.figure(figsize=(12,7))
total_wins.head(10).plot(kind='barh', color='skyblue')
plt.title('Top 10 Teams by Total Wins')
plt.xlabel('Total Wins')
plt.gca().invert_yaxis()
plt.show()
```
